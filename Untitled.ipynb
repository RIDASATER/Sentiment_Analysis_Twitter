```python
data_twitter_copy = data_twitter_copy.dropna()
data_twitter_copy.shape
```

```python
# Lowercasing
data_twitter_copy['clean_text'] = data_twitter_copy['clean_text'].str.lower()
data_twitter_copy['clean_text'].head()
```

```python
# Removing Punctuation & Special Characters
data_twitter_copy['clean_text'] = data_twitter_copy['clean_text'].astype(str).apply(lambda x: re.sub(r'[^\w\s]', '', x))
data_twitter_copy['clean_text']
```

```python
# Removal stop_words
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stop_words(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

data_twitter_copy['clean_text'] = data_twitter_copy['clean_text'].apply(remove_stop_words)
data_twitter_copy['clean_text'].head()
```

```python
# Removal of URLS
def remove_urls(text):
    url_pattern = re.compile(r'https?://S+|www\.\S+')
    return url_pattern.sub(r'', text)

data_twitter_copy['clean_text'] = data_twitter_copy['clean_text'].apply(remove_urls)
data_twitter_copy['clean_text'].head()
```

```python
# Romoval html Tags
def remove_html_tags(text):
    html_tags_pattern = r'<.*?>'
    return re.sub(html_tags_pattern, '', text)
data_twitter_copy['clean_text'] = data_twitter_copy['clean_text'].apply(remove_html_tags)
data_twitter_copy['clean_text'].head()
```

```python
data_twitter_copy['category']=data_twitter_copy['category'].apply(lambda x: "postive" if x==1 else "Negative" if x==-1 else "Neutral")
```

```python
X_train , X_test, y_train, y_test = train_test_split(data_twitter_copy['clean_text'], data_twitter_copy['category'],
                                                    test_size = 0.2 , random_state=0)
```

```python
vectorizer = TfidfVectorizer()
X_train_vectors = vectorizer.fit_transform(X_train)
X_test_vectors = vectorizer.transform(X_test)
```

```python
# Initialize Logistic Regression model
model = LogisticRegression()
# Fit the model on the training data
model.fit(X_train_vectors, y_train)
```

```python
import pickle
# Save the model to a file using 'wb' (write binary) mode
pickle.dump(model, open("twitter_sentiment.pkl" , 'wb'))
```
